# Correlation Analysis

```python
# correlation matrix of stocks
df.corr()
```

```python
first_10_stocks = df.columns[1:11]
returns_pivot =   df[first_10_stocks].pct_change().dropna()
correlation_matrix = returns_pivot.corr()
print(correlation_matrix)

correlation_pairs = correlation_matrix.unstack().sort_values(kind="quicksort", ascending=False)

most_correlated_pairs = correlation_pairs[correlation_pairs < 1].drop_duplicates()

print(most_correlated_pairs.head(10))

plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Stock Returns')
plt.show()
```

# Time Series Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

df.set_index('Date', inplace=True)

chosen_stock = df['TCS'] 

chosen_stock = chosen_stock.dropna()

result = seasonal_decompose(chosen_stock, model='multiplicative', period=30)

plt.figure(figsize=(14, 10))
plt.subplot(411)
plt.plot(result.observed, label='Observed')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(result.trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(result.seasonal, label='Seasonality')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(result.resid, label='Residuals')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

print("Trend component:")
print(result.trend.dropna().head())

print("Seasonality component:")
print(result.seasonal.dropna().head())

print("Residual component:")
print(result.resid.dropna().head())
```

# Anomaly Detection

```python
returns = df.pct_change(fill_method=None)

z_scores = (returns - returns.mean()) / returns.std()

threshold = 3

anomalies = (z_scores > threshold) | (z_scores < -threshold)

top_anomalies = anomalies.sum().nlargest(5)
print("Top 5 most significant anomalies:")
print(top_anomalies)

for stock, count in top_anomalies.items():
    print(f"\nAnomalies for {stock}:")
    print(df[stock][anomalies[stock]])
```
